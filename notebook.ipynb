```python
# Add your code snippets here
print(tweets.columns)
#tweets['timestamp_ms'].head()
def tweets_at_timestamp(timestamp):
    '''Returns the text of any tweet(s) published at a specific timestamp'''
    matching_tweets = tweets[tweets['timestamp_ms'] == timestamp]['text']
    return matching_tweets

x = tweets_at_timestamp(1459294817758)
print(x)

def get_tweets_by_timestamp(timestamp, *args):
    # Default to "text" column if no additional columns are specified
    columns = args if args else ["text"]
    print(columns)
    
    # Filter the DataFrame for the specified timestamp and select the specified columns
    matching_tweets = tweets[tweets['timestamp_ms'] == timestamp][list(columns)]
    return matching_tweets

result = get_tweets_by_timestamp(1459294817758, "created_at", "lang")
print(result)
```
